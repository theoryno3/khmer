```python
import mdp

def do_pca2(d):
    data = numpy.array(d)
    means = numpy.mean(data, axis=0)
    std = numpy.std(data, axis=0)

    for j in range(data.shape[1]):
        if std[j] != 0:
            data[:,j] = (data[:,j] - means[j]) / std[j]
        
    pcan = mdp.nodes.PCANode(output_dim=5)
    pcar = pcan.execute(data)
        
    return pcar, pcan

def do_pca(d):
    pcar, pcan = do_pca2(d)
    return pcar

# project does a PCA on orig, and then moves 'data' into that set of principle components
# without doing a PCA on data at all.
def project(data, orig):
    assert data.shape[1] == orig.shape[1]
    
    orig = numpy.array(orig)
    means = numpy.mean(orig, axis=0)
    std = numpy.std(orig, axis=0)
    
    data = numpy.array(data)

    for j in range(orig.shape[1]):
        if std[j] != 0:
            orig[:,j] = (orig[:,j] - means[j]) / std[j]
            data[:,j] = (data[:,j] - means[j]) / std[j]

    pcan = mdp.nodes.PCANode(output_dim=5)
    pcan.execute(orig)
    
    pcar = pcan.execute(data)
        
    return pcar
```

```python
# sA_contigs0205 is Hana's CLC assembly of QP's diginorm + filter-below-abund reads.
sA_contigs0205_pca, sA_contigs0205_node = do_pca2(sA_contigs0205)
```
